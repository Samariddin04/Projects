```python
df_pop = df[((df['CBSA'] == 'Los Angeles-Long Beach-Anaheim CA') | (df['CBSA'] == 'New York-Newark-Jersey City NY-NJ-PA') |
          (df['CBSA'] == 'San Francisco-Oakland-Hayward CA') | (df['CBSA'] == 'Trenton NJ') | (df['CBSA'] == 'Urban Honolulu HI')
            | (df['Year'] == '1980') | (df['Year'] == '1985') | (df['Year'] == '1990') | (df['Year'] == '1995') | (df['Year'] == '2000')
            | (df['Year'] == '2005') | (df['Year'] == '2010') | (df['Year'] == '2015') | (df['Year'] == '2020'))]
```

```python
df_pop.head()
```

```python
import matplotlib.pyplot as plt
```

```python
data = []  
data.append(df_pop.loc[df_pop['CBSA'] == 'Los Angeles-Long Beach-Anaheim CA']['NO2AnnualMean'])
data.append(df_pop.loc[df_pop['CBSA'] == 'New York-Newark-Jersey City NY-NJ-PA']['NO2AnnualMean'])
data.append(df_pop.loc[df_pop['CBSA'] == 'San Francisco-Oakland-Hayward CA']['NO2AnnualMean'])
data.append(df_pop.loc[df_pop['CBSA'] == 'Trenton NJ']['NO2AnnualMean'])
data.append(df_pop.loc[df_pop['CBSA'] == 'Urban Honolulu HI']['NO2AnnualMean'])
data_min = min([min(sublist) for sublist in data]) 
data_max = max([max(sublist) for sublist in data]) 

fig, ax = plt.subplots()

fig.set_figheight(5)
fig.set_figwidth(7)
ax.boxplot(data)

ax.yaxis.axes.set_ylim(data_min - 2, data_max + 2)
ax.set_xticklabels(['Los Angeles','New York', 'San Francisco',
                   'Trenton', 'Honolulu'])
ax.set_ylabel('NO2 Annual Mean')

ax.xaxis.set_tick_params(which = 'both', top = False, bottom = True, labelbottom = True) 
ax.yaxis.set_tick_params(which = 'both', right = False, left = True, labelleft = True)   
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
```

```python
fig, ax = plt.subplots()
ax.plot(df_pop[df_pop['CBSA'] == 'Los Angeles-Long Beach-Anaheim CA'].Year, df_pop[df_pop['CBSA'] == 'Los Angeles-Long Beach-Anaheim CA'].NO2AnnualMean, color = 'green', label = 'Los Angeles')
ax.plot(df_pop[df_pop['CBSA'] == 'New York-Newark-Jersey City NY-NJ-PA'].Year, df_pop[df_pop['CBSA'] == 'New York-Newark-Jersey City NY-NJ-PA'].NO2AnnualMean, color = 'blue', label = 'New York')
ax.plot(df_pop[df_pop['CBSA'] == 'San Francisco-Oakland-Hayward CA'].Year, df_pop[df_pop['CBSA'] == 'San Francisco-Oakland-Hayward CA'].NO2AnnualMean, color = 'orange', label = 'San Francisco')
ax.plot(df_pop[df_pop['CBSA'] == 'Trenton NJ'].Year, df_pop[df_pop['CBSA'] == 'Trenton NJ'].NO2AnnualMean, color = '#CA02CA', label = 'Trenton')
ax.plot(df_pop[df_pop['CBSA'] == 'Urban Honolulu HI'].Year, df_pop[df_pop['CBSA'] == 'Urban Honolulu HI'].NO2AnnualMean, color = 'red', label = 'Honolulu')

ax.set_xlim(1980,2020)
fig.set_size_inches(12,8)
fig.suptitle('The Annual Mean of NO2 Over Time', fontsize=18)
ax.legend(loc = 1, fontsize = 12)
ax.xaxis.set_label_text('Year')
ax.yaxis.set_label_text('NO2 Annual Mean')
plt.show()
```
