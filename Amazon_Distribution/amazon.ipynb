```python
import gurobipy as gpy
import numpy as np

''' Import or define problem data '''
order = np.genfromtxt(r'C:\Users\ASUS\Desktop\Optimization\amazon_distribution_data\order_100_100.txt')
tship = np.genfromtxt(r'C:\Users\ASUS\Desktop\optimization\amazon_distribution_data\tshp100_3_3.txt')
tship = tship.reshape(100,3,3)
inv = np.genfromtxt(r'C:\Users\ASUS\Desktop\optimization\amazon_distribution_data\inv100_100_3.txt')
cship = np.genfromtxt(r'C:\Users\ASUS\Desktop\optimization\amazon_distribution_data\ship_cost100_100_3.txt')
num_fc = cship.shape[1]
num_order = order.shape[1]

''' Create Gurobi model object '''
m = gpy.Model('amazon')

''' Specify whether model is maximized or minimized   (model sense) '''
m.ModelSense = gpy.GRB.MINIMIZE

''' Create decision variables and update model'''
tran_ship = m.addMVar((num_order,num_fc,num_fc), vtype=gpy.GRB.CONTINUOUS, name='x') 
daily_order = m.addMVar((num_order, num_fc), name='y')
order_fulfill = m.addMVar((num_order,num_fc), name='z', vtype=gpy.GRB.BINARY)
m.update()

''' Create objective function and update model '''
m.setObjective((tship*tran_ship).sum()+(cship*daily_order).sum())
m.update()

''' Create constraints and update model '''
m.addConstrs(inv + tran_ship.sum(axis=1) - tran_ship.sum(axis=2) - order.T@order_fulfill >= 0 for _ in range(1))
m.addConstrs(daily_order.sum(axis=0) <= 5000 for _ in range(1))

''' Optimize model '''
m.optimize()

''' Print results '''
for v in m.getVars():
    print(f'{v.varName}: {v.x}')
```
